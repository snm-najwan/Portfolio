```python
# first import relevant packages
import pandas as pd
import numpy as np
```

```python
# get the csv dataset and copy the original and peek
data = pd.read_csv("C:\\Users\\Najwan\\OneDrive\\Desktop\\Billionaires Statistics Dataset.csv")
dt = data.copy()
df = pd. DataFrame(dt)
df.head()
```

```python
# get the dimension of the data
df.shape
```

```python
# get all the column headers
list(df)
```

```python
# get a little more idea about the dataset
df.describe()
```

```python
# since row count is not same for every column check for null values
for i in list(df):
    print(i, " : ", df[i].isna().sum())
```

```python
# remove unwanted and columns with most null values
null_cols = ["state", "residenceStateRegion", "title", "organization", "birthYear", "birthMonth", "birthDay", "birthDate"]
df = df.drop(null_cols, axis = 1)
list(df)
```

```python
# check again for null values
for i in list(df):
    print(i, " : ", df[i].isna().sum())
```

```python
# check how many rows are with null values
nan_rows = df.isna().any(axis=1)
k=0
for i in range(0,len(nan_rows)):
    if nan_rows[i] == True:
        k+=1
print(k)
```

```python
# so totally 261 rows with null values, remove them
cl_df = df.dropna()
cl_df.head()
```

```python
cl_df.shape
```

```python
for i in list(df):
    print(i, " : ", df[i].dtype)
```

```python
cl_df.describe()
```

```python
# age, population cannot be float so change it into int
cl_df['age'] = cl_df['age'].astype(int)
cl_df['population_country'] = cl_df['population_country'].astype(int)

for i in list(cl_df):
    print(i, " : ", cl_df[i].dtype)
```

```python
cl_df.head()
```

```python
# since the ranking is wrong rerank
cl_df["rank"] = cl_df["finalWorth"].rank(ascending=False)
# cl_df["rank"].shape, cl_df["finalWorth"].shape
cl_df.head(20)
```
